# Analise exploratoria dos dados

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import process as p
```

```python
sample = pd.read_csv('train.csv', nrows = 5000000)
```

```python
sample.head()
```

```python
sample.info()
```

```python
sample = p.add_fields(sample)
```

```python
sample = p.clear_data(sample)
```

```python
fig, ax = plt.subplots()
plt.scatter(x = sample['distance'], y = sample['fare_amount'])
plt.grid()
ax.set_xlabel('distancia percorrida (em graus)')
ax.set_ylabel('valor pago')
```

```python
sample.groupby('ano').mean()
```

```python
sample.groupby('hora').mean().sort_values('fare_amount', ascending = False)
```

```python
sample.groupby('passenger_count').mean()
```

```python
fig, ax = plt.subplots()
plt.grid()
plt.hist(sample['fare_amount'], bins = 100)
ax.set_xlim(0,100)
```

```python
df_hora = sample.groupby('hora')
```

```python
fig, ax = plt.subplots()
plt.scatter(x = df_hora['hora'].count().keys(),y = df_hora.mean()['fare_amount'])
plt.grid()
ax.set_xlabel('hora')
ax.set_ylabel('fare_amount')
ax.set_title('grafico valor_pago x hora')
fig.set_size_inches(15, 8)
```

```python
df_ano = sample.groupby('ano')
```

```python
fig, ax = plt.subplots()
plt.scatter(x = df_ano['ano'].count().keys(),y = df_ano.mean()['fare_amount'])
plt.grid()
ax.set_xlabel('ano')
ax.set_ylabel('fare_amount')
ax.set_title('grafico valor_pago x ano')
fig.set_size_inches(15, 8)
```

```python
df_passenger = sample.groupby('passenger_count')
```

```python
fig, ax = plt.subplots()
plt.scatter(x = df_passenger['passenger_count'].count().keys(),y = df_passenger.mean()['fare_amount'])
plt.grid()
ax.set_xlabel('numero de passageiros')
ax.set_ylabel('fare_amount')
ax.set_title('grafico valor_pago x numero de passageiros')
fig.set_size_inches(15, 8)
```

```python
df_mes = sample.groupby('mes')
```

```python
fig, ax = plt.subplots()
plt.scatter(x = df_mes['mes'].count().keys(),y = df_mes.mean()['fare_amount'])
plt.grid()
ax.set_xlabel('mes')
ax.set_ylabel('fare_amount')
ax.set_title('grafico valor_pago x mes')
fig.set_size_inches(15, 8)
```
